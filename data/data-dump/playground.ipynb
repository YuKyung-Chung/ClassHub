```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import pandas as pd
import json

try:
    # MySQL 설정값
    host = 'localhost'
    port = 3306
    user = 'root'  # 데이터베이스 사용자 이름을 입력해주세요.
    password = '1234'  # 데이터베이스 비밀번호를 입력해주세요.
    db_name = 'classhub'
    # SQLAlchemy 엔진 객체 생성
    engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}/{db_name}")

    # Session 생성
    Session = sessionmaker(bind=engine)
    session = Session()
except:
    raise Exception("세션 연결 오류")

try:
    # CSV 파일 읽기
    category_df = pd.read_csv('./data/category.csv')  # CSV 파일 경로를 입력해주세요.

    # DataFrame을 MySQL에 덤프
    category_df.to_sql('category', engine, if_exists='append', index=False)
except:
    raise Exception("카테고리 삽입 오류")

try:
    # CSV 파일 읽기
    tag_df = pd.read_csv('./data/tag.csv')  # CSV 파일 경로를 입력해주세요.

    # DataFrame을 MySQL에 덤프
    tag_df.to_sql('tag', engine, if_exists='append', index=False)
except:
    raise Exception("태그 삽입 오류")

try:
    category_df = pd.read_sql("select category_id, category_name from category", engine)
    udemy_lectures = pd.read_csv("../udemy/lectures/udemy_lectures_dummy.csv")
    udemy_lectures = pd.merge(udemy_lectures, category_df, how='left', left_on='category_name', right_on='category_name')
    base_url = "https://storage.googleapis.com/classhub/data/udemy/htmlFiles/"
    # udemy_lectures[udemy_lectures["category_id"].isna()==True]
    udemy_lectures["description_detail"] = ["1.html"]*len(udemy_lectures)
    udemy_lectures["gpt_review_good"] = ["이 강의는 체계적이며 이해하기 쉽습니다.\n강의자를 신뢰할수있으며 도움이 많이 됩니다."]*len(udemy_lectures)
    udemy_lectures["gpt_review_bad"] = ["커리큘럼이 다른 강의에 비해 조금 긴편이며 강의를 보는것외에 실습에 시간을 꽤 소요해야합니다."]*len(udemy_lectures)
    udemy_lectures['curriculum'] = udemy_lectures['curriculum'].apply(json.dumps)
    udemy_lectures = udemy_lectures.drop(columns=["category_name"])
    print(udemy_lectures.columns)
except:
    raise Exception("강의 테이블 데이터 생성 실패")


try:
    # 강의 테이블 덤프
    udemy_lectures.to_sql('lecture', engine, if_exists='append',
                          method='multi', index=False)
    

except Exception as e:
    # 에러 발생 시 롤백
    session.rollback()
    print(f"Error occurred: {e}")
    raise Exception("강의 테이블 덤프 실패")

finally:
    # Session 종료
    print(pd.read_sql('SELECT * FROM lecture limit 100', engine))
    session.close()

try:
    udemy_lectures = pd.read_sql('SELECT lecture_id, site_lecture_id FROM lecture', engine)

    udemy_tags = pd.read_csv("../data-dump/udemy_lecture_cat_tags.csv")[["id","tags"]]
    # print(udemy_tags)

    tag_table = pd.read_sql("select tag_id, name from tag", engine)
    # print(tag_table)
    
    tag_merged = pd.merge(udemy_tags, tag_table, how='left', left_on='tags', right_on='name')
    tag_merged["id"] = tag_merged["id"].apply(lambda x:f"u{x}")
    # print(tag_merged)

    # tag_merged
    lecture_tag = pd.merge(udemy_lectures, tag_merged, how='inner', left_on='site_lecture_id', right_on='id')
    lecture_tag = lecture_tag[["lecture_id","tag_id"]]
    # print(lecture_tag[["lecture_id","tag_id"]].isnull().sum())

    # 강의 테이블 덤프
    lecture_tag.to_sql('lecture_tag', engine, if_exists='append',
                          method='multi', index=False)
    
    session.rollback()
except:
    raise Exception("lecture_tag 삽입 실패")


    # # 트랜잭션 커밋
#     session.commit()
# except Exception as e:
#     # 에러 발생 시 롤백
#     session.rollback()
#     print(f"Error occurred: {e}")
#     raise Exception("강의 테이블 덤프 실패")

# finally:
#     # Session 종료
#     print(pd.read_sql('SELECT * FROM lecture limit 100', engine))
#     session.close()
```

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import pandas as pd
import json

try:
    # MySQL 설정값
    host = 'localhost'
    port = 3306
    user = 'root'  # 데이터베이스 사용자 이름을 입력해주세요.
    password = '1234'  # 데이터베이스 비밀번호를 입력해주세요.
    db_name = 'classhub'
    # SQLAlchemy 엔진 객체 생성
    engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}/{db_name}")

    # Session 생성
    Session = sessionmaker(bind=engine)
    session = Session()
except:
    raise Exception("세션 연결 오류")
try:
    # CSV 파일 읽기
    category_df = pd.read_csv('./data/category.csv')  # CSV 파일 경로를 입력해주세요.

    # DataFrame을 MySQL에 덤프
    category_df.to_sql('category', engine, if_exists='append', index=False)
except:
    session.rollback()

    raise Exception("카테고리 삽입 오류")

try:
    # CSV 파일 읽기
    tag_df = pd.read_csv('./data/tag.csv')  # CSV 파일 경로를 입력해주세요.

    # DataFrame을 MySQL에 덤프
    tag_df.to_sql('tag', engine, if_exists='append', index=False)
except:
    session.rollback()

    raise Exception("태그 삽입 오류")

try:
    category_df = pd.read_sql("select category_id, category_name from category", engine)
    udemy_lectures = pd.read_csv("../udemy/lectures/udemy_lectures_dummy.csv")
    udemy_lectures = pd.merge(udemy_lectures, category_df, how='left', left_on='category_name', right_on='category_name')
    base_url = "https://storage.googleapis.com/classhub/data/udemy/htmlFiles/"
    # udemy_lectures[udemy_lectures["category_id"].isna()==True]
    udemy_lectures["description_detail"] = ["1.html"]*len(udemy_lectures)
    udemy_lectures["gpt_review_good"] = ["이 강의는 체계적이며 이해하기 쉽습니다.\n강의자를 신뢰할수있으며 도움이 많이 됩니다."]*len(udemy_lectures)
    udemy_lectures["gpt_review_bad"] = ["커리큘럼이 다른 강의에 비해 조금 긴편이며 강의를 보는것외에 실습에 시간을 꽤 소요해야합니다."]*len(udemy_lectures)
    udemy_lectures['curriculum'] = udemy_lectures['curriculum'].apply(json.dumps)
    udemy_lectures = udemy_lectures.drop(columns=["category_name"])
    print(udemy_lectures.columns)
except:
    session.rollback()

    raise Exception("강의 테이블 데이터 생성 실패")


try:
    # 강의 테이블 덤프
    udemy_lectures.to_sql('lecture', engine, if_exists='append',
                          method='multi', index=False)
    

except Exception as e:
    # 에러 발생 시 롤백
    session.rollback()
    print(f"Error occurred: {e}")
    raise Exception("강의 테이블 덤프 실패")

finally:
    # Session 종료
    print(pd.read_sql('SELECT * FROM lecture limit 100', engine))
    # session.close()

try:
    udemy_lectures = pd.read_sql('SELECT lecture_id, site_lecture_id FROM lecture', engine)

    udemy_tags = pd.read_csv("../data-dump/udemy_lecture_cat_tags.csv")[["id","tags"]]
    # print(udemy_tags)

    tag_table = pd.read_sql("select tag_id, name from tag", engine)
    # print(tag_table)
    
    tag_merged = pd.merge(udemy_tags, tag_table, how='left', left_on='tags', right_on='name')
    tag_merged["id"] = tag_merged["id"].apply(lambda x:f"u{x}")
    # print(tag_merged)

    # tag_merged
    lecture_tag = pd.merge(udemy_lectures, tag_merged, how='inner', left_on='site_lecture_id', right_on='id')
    lecture_tag = lecture_tag[["lecture_id","tag_id"]]
    # print(lecture_tag[["lecture_id","tag_id"]].isnull().sum())

    # 강의 테이블 덤프
    lecture_tag.to_sql('lecture_tag', engine, if_exists='append',
                          method='multi', index=False)
    
except:
    session.rollback()

    raise Exception("lecture_tag 삽입 실패")

try:
    review_df = pd.read_csv("../data-dump/udemy_review.csv")
    udemy_lectures = pd.read_sql('SELECT lecture_id, site_lecture_id FROM lecture', engine)
    review_df = pd.merge(left=review_df,right=udemy_lectures,how="left",left_on="lecture_id",right_on="site_lecture_id")
    review_df = review_df.drop(columns="lecture_id_x")
    review_df.rename(columns = {'lecture_id_y' : 'lecture_id'}, inplace = True)
    review_df.rename(columns = {'rating' : 'score'}, inplace = True)
    review_df = review_df.drop(columns="site_lecture_id")
    review_df['content'] = review_df['content'].str.slice(start=0, stop=300)
    review_df['content'] = review_df['content'].fillna("")
    review_df.to_sql('site_review', engine, if_exists='append',
                            method='multi', index=False)
except:
    session.rollback()

    raise Exception("site_review 삽입 실패")


```
